```python
import pandas as pd
import math
import numpy as np
import jellyfish
```

```python
# Getting the unigram text data from the link
!wget https://norvig.com/ngrams/count_1w.txt
# Updating the unigram data into dataframe using pandas
unigrams_df = pd.read_csv('count_1w.txt' , delimiter = '\t', names = ['unigram','frequency'])
unigrams_df.head(100)
```

```python
total_count = unigrams_df['frequency'].sum()
total_count
```

```python

def nonWordSpellCorrection(nonWord):
  print(f'the input is : {nonWord}')
  words = []
  for row in range(len(unigrams_df['unigram'])):

    dist = jellyfish.damerau_levenshtein_distance(str(unigrams_df['unigram'][row]), nonWord)

    if dist == 1:
      unigram = [unigrams_df.iloc[row][0], unigrams_df.iloc[row][1], unigrams_df.iloc[row][1]/total_count]
      words.append(unigram)

  new_correctedata = pd.DataFrame(words, columns =['unigram', 'Frequency', 'P(word)'])

  return new_correctedata
```

```python
nonWordSpellCorrection('acress')
```

```python
from textblob import TextBlob, Word

candidates = []
def word_check(word):
    word = Word(word)
    result = word.spellcheck()
    candidates.append(result)
    print(f'Candidates = {candidates}')

word_check('acress')
```

```python
def sentence_check(sentence):
    print(f'The input sentence is: {sentence}')
    sentence = TextBlob(sentence)

    result = sentence.correct()

    print(f'The corrected sentence is: {result}')


sentence_check('A sentencee to checkk!')
```
