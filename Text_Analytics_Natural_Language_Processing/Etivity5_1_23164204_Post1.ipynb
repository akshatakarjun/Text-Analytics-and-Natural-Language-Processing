```python
import pandas as pd

!wget http://ptrckprry.com/course/ssd/data/positive-words.txt

positive_df = pd.read_csv('positive-words.txt', delimiter = '\t', header = None)
positive_df
```

```python
!wget http://ptrckprry.com/course/ssd/data/negative-words.txt

negative_df = pd.read_csv('negative-words.txt', delimiter = '\t', header = None, encoding="ISO-8859-1")
negative_df
```

```python
def sentimentAnalyzer(string):
  list_words = string.split(' ')

  positive_words = 0
  negative_words = 0
  total_words = len(list_words)

  for x in positive_df.itertuples():
    for y in list_words:
      if y == x[1]:
        positive_words +=1

  for x in negative_df.itertuples():
    for y in list_words:
      if y == x[1]:
        negative_words +=1


  print(f'string = {string}')
  if positive_words > negative_words:
    print('positive sentiment')
    print(f'confidence : {positive_words/total_words}' )
  elif positive_words < negative_words:
    print('negative_sentiment')
    print(f'confidence : {negative_words/total_words}' )
  else:
    print('neutral sentiment')
    print(positive_words - negative_words)
  print('--------------')
```

```python
sentimentAnalyzer('NLP is cool')
sentimentAnalyzer('NLP is cool and useful')
sentimentAnalyzer('NLP is hard')
sentimentAnalyzer('NLP is hard and useless')
sentimentAnalyzer('NLP stands for Natural Language Processing')
```
