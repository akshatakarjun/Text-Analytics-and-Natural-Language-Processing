Probabilty of sentence without laplace smoothing

```python
def probability(sentence):
    #inititalizing the porbability of the sentence as 1
    sentenceProbability = 1
    #Spliiting the sentence into words
    words = sentence.lower().split()
    print(f"Sentence: {words}")

    #Looking for the first(unigram) and second_word's(bigram) count from the list of words
    for word, second_word in zip(words, words[1:]):
        unigram_count = unigrams_df.loc[unigrams_df['unigram'] == word]['count'].values[0]
        print(f"Unigram count for {word}: {unigram_count}")
        bigram = f"{word} {second_word}"
        bigram_count = bigrams_df.loc[bigrams_df['bigram'] == bigram]['count'].values[0]
        print(f"Bigram count for {bigram}: {bigram_count}")

        #Calculating the bigram probability
        bigramProbability = bigram_count/unigram_count
        print(f"Bigram probability for {word} {second_word}: {bigramProbability}")
        print("-"*80)

        #Updating the probability of sentence by multiplying with probability of bigram after each iteration of word in the sentence
        sentenceProbability *= bigramProbability

    return sentenceProbability
```

```python
probability("i love you")>probability('i hate you')
```

Probability of sentence with laplace smoothing

```python
def probability(sentence):

    sentenceProbability = 0
    words = sentence.lower().split()
    print(f"Sentence: {words}")

    for word, second_word in zip(words, words[1:]):
        unigram_count = unigrams_df.loc[unigrams_df['unigram'] == word]['count'].values[0]
        print(f"Unigram count for {word}: {unigram_count}")
        bigram = f"{word} {second_word}"
        bigram_count = bigrams_df.loc[bigrams_df['bigram'] == bigram]['count'].values[0]
        print(f"Bigram count for {bigram}: {bigram_count}")

        #Taking the log of probability to avoid the probability of a word being 0
        bigramProbability = math.log((bigram_count+1)/(unigram_count+totalUnigrams))
        print(f"Bigram probability for {word} {second_word}: {math.exp(bigramProbability)}")
        print("-"*80)
        #Adding the bigram probability instead of multplying since we have applied laplace smoothing to avoid the error of 0/inf
        sentenceProbability += bigramProbability


    return math.exp(sentenceProbability)
```

```python
probability("i love you")>probability('i hate you')
```

```python
def ShannonVisualisation(seed="<S>", max_length=100):

    #Calculating the bigram probability from the bigrams dataframe
    bigramProbability = {}
    for _, row in bigrams_df.iterrows():

        #Extracting the bigrama and its count from each row in bigrams dataframe
        bigram, count = row['bigram'], row['count']
        first_word, second_word = bigram.split()
        if first_word not in bigramProbability:
            bigramProbability[first_word] = {second_word: count}
        else:
            bigramProbability[first_word][second_word] = bigramProbability[first_word].get(second_word, 0) + count

    #Converting the counts to probabilities
    for first_word in bigramProbability:
        total_count = sum(bigramProbability[first_word].values())
        for second_word in bigramProbability[first_word]:
            bigramProbability[first_word][second_word] /= total_count

    #Generating text
    current_word = seed
    generated_text = []
    while current_word != "</S>" and len(generated_text) < max_length:
        next_words_probs = bigramProbability.get(current_word, None)
        print("Current word: " + current_word + "\n")

        #Break if no next words are available
        if not next_words_probs:
            break

        words, probabilities = zip(*next_words_probs.items())
        current_word = np.random.choice(words, p=probabilities)
        generated_text.append(current_word)

    return ' '.join(generated_text).strip()
```

```python
max_sent_length = input("How many words in a sentence ")

generated_text = ShannonVisualisation(seed="<S>", max_length=int(max_sent_length))
print(generated_text)
```
